```python
import autokeras as ak
import pandas as pd
import numpy as np

platforms = ["Pitchfork", "Guardian", "NME", "Spectrum"]
dfs = {platform: 
       pd.read_hdf("C:\\Users\\tommy\\OneDrive\\University\\Year 3\\Third Year Project\\Platform Album Data\\text_edited_data.h5", key=platform) for platform in platforms}
```

```python
import math

dfs["Pitchfork"].loc[:, "Scores"] = (dfs["Pitchfork"].loc[:, "Scores"].apply(float).div(2)).apply(math.ceil)
dfs["Spectrum"].loc[:, "Scores"] = (dfs["Spectrum"].loc[:, "Scores"].apply(float) / 20).apply(math.ceil)

print(dfs)
```

```python
text = []
scores = []

for df in list(dfs.values())[:3]:
    
    text += list(df.loc[:, "Text"].values)
    scores += list(df.loc[:, "Scores"].values)
    
text += list(dfs["Spectrum"].loc[dfs["Spectrum"]["Scores"] != 0, "Text"])
scores += list(dfs["Spectrum"].loc[dfs["Spectrum"]["Scores"] != 0, "Scores"])

text = np.array(text)
scores = np.array(scores)
```

```python
reg = ak.TextRegressor(overwrite=True, max_trials=1)

reg.fit(text, scores)
preds = reg.predict(text)

print(reg.evaluate(scores, preds))
```
