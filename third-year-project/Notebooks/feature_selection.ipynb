```python
correlation_matrix = X.corr()
correlation_matrix = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape)).astype(np.bool_))
```

```python
correlation_matrix = correlation_matrix.stack().reset_index().rename({"level_0": "var1", "level_1": "var2", 0: "Correlation"}, axis=1)
correlation_matrix = correlation_matrix.loc[correlation_matrix["var1"] != correlation_matrix["var2"]]
```
